```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm

df = pd.read_csv ("sub-data.csv")
df.head()
```

```python
x = "ASSET"
Y = "DEBT"
```

```python
plt.scatter(x = df[x], y= df[Y])
plt.title("Assets in Comparisson to debt")  
```

```python
x_ary = df["ASSET"]
x_with_cnst = sm.add_constant(x_ary)
model = sm.OLS(df["DEBT"], x_with_cnst)
results = model.fit()
b, m = results.params
print(m)
print(b)
```

```python
print(results.summary())
```

```python
plt.figure(figsize = (20,20))
plt.scatter(df["ASSET"], df["DEBT"])
plt.plot(x_ary, m*x_ary+b, color = "orange")
```
